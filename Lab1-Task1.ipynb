```python
import matplotlib.pyplot as plt

# Generate arrays
exp_data = np.random.exponential(scale=1.0, size=(100000,))
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

# Plot histograms
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="green", label="Normal")
plt.hist(exp_data, density=True, bins=100, histtype="step", color="red", label="Exponential")

# Adjust plot
plt.axis([-2.5, 8, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Create grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Labels and title
ax.set_title("Z = X² + Y²")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Select the relevant features
features = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
selected_df = df[features].dropna()

# Calculate Pearson correlation
pearson_corr = selected_df.corr(method='pearson')

# Calculate Spearman correlation
spearman_corr = selected_df.corr(method='spearman')

# Set up the plot
plt.figure(figsize=(14, 6))

# Pearson heatmap
plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Pearson Correlation")

# Spearman heatmap
plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, fmt=".2f", cmap="YlGnBu", square=True)
plt.title("Spearman Correlation")

plt.tight_layout()
plt.show()
```
